```python
#dados de interesse

df = receitas_deputados_eleitos[['Sequencial Candidato','Nome candidato', 'Nome do doador', 'CPF/CNPJ do doador', 'Valor receita', 'Nome do doador originário', 'CPF/CNPJ do doador originário']]
```

```python
#iteração PARA PREENCHER DOADOR ORIGINARIO SE O CAMPO ESTIVER #NULO - neste caso, como o próprio doador é o doador originário, preencheremos os dados do doador originário com os dados do próprio doador.

for index, row in df.iterrows():
  if row['CPF/CNPJ do doador originário'] == '#NULO': 
      if row['CPF/CNPJ do doador'] != '#NULO':
        df.at[index,'CPF/CNPJ do doador originário'] = row['CPF/CNPJ do doador']
        df.at[index,'Nome do doador originário'] = row['Nome do doador']
```

```python
#Checando se ainda existem nulos 

df[df['CPF/CNPJ do doador originário'] == '#NULO']

```

```python
#conhecendo as doações irrastreáveis
df[df['CPF/CNPJ do doador originário'] == '#NULO'][['Nome candidato', 'Valor receita']].groupby('Nome candidato').sum('Valor receita').sort_values(['Valor receita'], ascending=False)
```

```python
#136 doações para 47 deputados irrastreáveis, linhas serão descartadas
df = df.drop(df[df['CPF/CNPJ do doador originário'] == '#NULO'].index)
```

```python
#colocar CPF/CNPJ como número, minimizando duplicidades

df['CPF/CNPJ do doador originário'] = df['CPF/CNPJ do doador originário'].astype(int)
```

```python
#eliminar todas as pessoas físicas da tabela - devolver CNPJ para string

df['CPF/CNPJ do doador originário'] = df['CPF/CNPJ do doador originário'].astype(str)
df = df[df['CPF/CNPJ do doador originário'].str.match(r'(.*0001\d{2})') == True ]
```

```python
#padronizar os CNPJs com 14 digitos
for index, row in df.iterrows():
  if len(row['CPF/CNPJ do doador originário']) < 14:
    dif = 14 - len(row['CPF/CNPJ do doador originário'])
    tam = dif*'0' 
    df.at[index,'CPF/CNPJ do doador originário'] = (tam + row['CPF/CNPJ do doador originário'])
   
```

```python
#analisando os maiores doadores 

df[['CPF/CNPJ do doador originário', 'Nome do doador originário', 'Valor receita']].groupby(['CPF/CNPJ do doador originário', 'Nome do doador originário']).sum('Valor receita').sort_values(['Valor receita'], ascending=False)
```

```python
#problema: diversos nomes diferentes para o mesmo CPF/CNPJ
#solucao - criar funcao que, para cada CPF/CNPJ, substitua a ocorrência do nome pela primeira ocorrencia do nome, uniformizando o nome associado ao CNPJ/CPF


def substituir_CPF_CNPJ(df):
    """
    Essa função irá substituir o nome da empresa/pessoa pela sua primeira ocorrência no dataframe
    """
    CPF_CNPJ = {}
    
    for index,row in df.iterrows():
        if row['CPF/CNPJ do doador originário'] in CPF_CNPJ.keys():
           df.at[index,'Nome do doador originário'] = CPF_CNPJ[row['CPF/CNPJ do doador originário']]
        else:
           CPF_CNPJ[row['CPF/CNPJ do doador originário']] = row['Nome do doador originário']    




```

```python
#tratar toda a base de dados
substituir_CPF_CNPJ(df)
```

```python
#retirar doações dos CNPJs dos FUNDOS PARTIDÁRIOS
df = df.drop(df[df['CPF/CNPJ do doador originário'].isin(['08517423000195', '01633510000169','00887169000105','01421697000137', '04514550000180', '19943073000145', '00719575000169', '19943073000145', '24472136000162', r'(2056.*)', r'(2055.*)', r'(2057.*)', r'(2063.*)', r'(2060.*)' ])].index)
df = df.drop(df[df['CPF/CNPJ do doador originário'].str.match(r'(2059.*)') == True ].index)
df = df.drop(df[df['CPF/CNPJ do doador originário'].str.match(r'(2056.*)') == True ].index)
df = df.drop(df[df['CPF/CNPJ do doador originário'].str.match(r'(2057.*)') == True ].index)
df = df.drop(df[df['CPF/CNPJ do doador originário'].str.match(r'(2055.*)') == True ].index)
df = df.drop(df[df['CPF/CNPJ do doador originário'].str.match(r'(2060.*)') == True ].index)
df = df.drop(df[df['CPF/CNPJ do doador originário'].str.match(r'(2063.*)') == True ].index)
df = df.drop(df[df['Nome do doador originário'].str.match(r'(DIRETORIO.*)') == True ].index)
df = df.drop(df[df['Nome do doador originário'].str.match(r'(PARTIDO.*)') == True ].index)
df = df.drop(df[df['Nome do doador originário'].isin(['DIREÇÃO NACIONAL', 'Direção Nacional','Direção Estadual/Distrital', 'Comitê Financeiro Único', 'Direção Municipal'])].index)
```

```python
# PROBLEMA: 

#queiroz galvão tem 4 empresas, bradesco tem 6 empresas

#solucao: padronizar o CNPJ delas e chamar de "GRUPO"

def substituir_CPF_CNPJ_BRADESCO(df):
    """
    Essa função irá substituir o nome e o CNPJ das empresas Bradesco pelo GRUPO BRADESCO
    """
    CPF_CNPJ = ['51990695000137', '92693118000160','47509120000182','52568821000122', '33010851000174', '61855045000132']
    
    for index,row in df.iterrows():
        if row['CPF/CNPJ do doador originário'] in CPF_CNPJ:
           df.at[index,'Nome do doador originário'] = "GRUPO BRADESCO"
           df.at[index,'CPF/CNPJ do doador originário'] = '51990695000137'
          

def substituir_CPF_CNPJ_QUEIROZ_GALVAO(df):
    """
    Essa função irá substituir o nome e o CNPJ das empresas Bradesco pelo GRUPO QUEIROZ GALVÃO
    """
    CPF_CNPJ = ['33412792000160', '04899037000154','11862944000194','08805525000106']
    
    for index,row in df.iterrows():
        if row['CPF/CNPJ do doador originário'] in CPF_CNPJ:
           df.at[index,'Nome do doador originário'] = "GRUPO QUEIROZ GALVÃO"
           df.at[index,'CPF/CNPJ do doador originário'] = '33412792000160'

```

```python
#deputados - preparando o arquivo de nós para o Gephi 
nos1 = df2[['Sequencial Candidato', 'NM_URNA_CANDIDATO', 'SG_PARTIDO', 'SG_UF']].drop_duplicates()
nos3 = df2[['Sequencial Candidato', 'Valor receita']].groupby('Sequencial Candidato').sum('Valor receita')
#merge dos candidatos com os totais recebidos
nos1 = nos1.merge(nos3, how='inner', on='Sequencial Candidato')
nos1.rename(columns={'Sequencial Candidato': 'Id','NM_URNA_CANDIDATO': 'Label' }, inplace = True)
nos1 = nos1.assign(Type=0)
nos1
```

```python
#TRANSITIVIDADE - autojoin - para criar relacionamentos entre deputados do mesmo partido e estado
nos7 = df2[['Sequencial Candidato', 'NM_URNA_CANDIDATO', 'SG_PARTIDO', 'SG_UF']].drop_duplicates()
nos8 = df2[['Sequencial Candidato', 'NM_URNA_CANDIDATO', 'SG_PARTIDO', 'SG_UF']].drop_duplicates()
nos8.rename(columns={'Sequencial Candidato': 'Sequencial Candidato2','NM_URNA_CANDIDATO': 'NM_URNA_CANDIDATO2' }, inplace = True)

nos9 = pd.merge(nos7,nos8, how='left', on = ["SG_PARTIDO","SG_UF"])

nos9
```

```python
#problema: relacionamentos repetidos
#solucao - criar funcao que encontra e remove os relacionamentos repetidos


def remover_relacionamentos_repetidos(df):
    """
    Essa função irá remover os relacionamentos repetidos
    """
    repetidos = []
    for index,row in df.iterrows():
      if row['Sequencial Candidato'] == row['Sequencial Candidato2']:
        repetidos.append(index)
           
    df.drop(index=repetidos, inplace=True)   
```

```python
#remoção dos relacionamentos repetidos
remover_relacionamentos_repetidos(nos9)
nos9
```
